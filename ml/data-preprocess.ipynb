```python
import pandas as pd

data = {
    "Name" : ["Alice", "Bob", "Charlie", "David"],
    "Age" : [24, 30, None, 35],
    "Salary" : [50000, None, 55000, None]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
total_missing_val = df.isnull().sum()
# to check how much % data is missing
print(df.isnull().mean() * 100)
```

```python
fill_with_0 = df.fillna(0)


drop_missing = df.dropna()


fill_with_mean = df.copy()
fill_with_mean['Age'].fillna(fill_with_mean['Age'].mean(), inplace=True)
fill_with_mean['Salary'].fillna(fill_with_mean['Salary'].mean(), inplace=True)
print("Fill with 0:\n", fill_with_0 , "\n\n\n")
print("Drop missing:\n", drop_missing , "\n\n\n")
print("Fill with mean:\n", fill_with_mean)
```
