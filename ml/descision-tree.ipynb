```python
from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

X_data = [
    [7,1],
    [6,3],
    [5,1],
    [8,7],
    [10,8],
    [9,9]
]
y_labels = [0, 0, 0, 1, 1, 1] # 0: Apple, 1: Orange

# Train the classifier
clf.fit(X_data, y_labels)


height = float(input("Enter fruit height: "))
shade = float(input("Enter fruit shade: "))

# Make a prediction
prediction = clf.predict([[height, shade]])

if prediction[0] == 0:
    print("The fruit is an Apple.")
else:
    print("The fruit is an Orange.")
```
