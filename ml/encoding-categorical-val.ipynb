```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

data = pd.read_csv("sample_data.csv")
df_labeled = data.copy()

label_encoder = LabelEncoder()
df_labeled["Gender_Encoded"] = label_encoder.fit_transform(df_labeled["Gender"])
df_labeled["Passed_Encoded"] = label_encoder.fit_transform(df_labeled["Passed"])

print(df_labeled.head(5))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split

data = {
    "Study_Hours": [2, 4, 6, 8, 10],
    "Scores": [50, 60, 70, 80, 90]
}

df = pd.DataFrame(data)

standard_scaler = StandardScaler()
df_standard_scaled = standard_scaler.fit_transform(df)

print("Standard Scaled Data:\n", pd.DataFrame(df_standard_scaled,columns=["Study_Hours","Scores"]),"\n\n\n")


minmax_scaler = MinMaxScaler()
df_minmax_scaled = minmax_scaler.fit_transform(df)

print("Min-Max Scaled Data:\n", pd.DataFrame(df_minmax_scaled,columns=["Study_Hours","Scores"]))


X_train, X_test, y_train, y_test = train_test_split(
    df[["Study_Hours"]],
    df["Scores"],
    test_size=0.2,
    random_state=42
)

print("\n\nX_train:\n", X_train)
print("\nX_test:\n", X_test)
print("\ny_train:\n", y_train)
print("\ny_test:\n", y_test)
```
