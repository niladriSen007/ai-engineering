```python
# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)

print(df)
```

```python
print(df.describe())
```

```python
ages = df['Age']
print(ages)
```

```python
subset = df[['Name', 'City']]
print(subset)
```

```python
filtered = df[df['Age'] > 28]
print(filtered)
```

```python
df['Country'] = 'USA'
print(df)
```

```python
df.rename(columns={'Name': 'First Name', 'City': 'Location'}, inplace=True)
print(df)
```

```python
# This will give you a count of missing values per column.
print(df.isnull().sum())
```

```python
# replace missing values with a specified value:
df['Age'] = df['Age'].fillna(0)
# if you want to forward-fill values,
df['Age'] = df['Age'].ffill()
# or backward-fill values
df['Age'] = df['Age'].bfill()
# drop rows with missing values
df = df.dropna()
```

```python
df['Salary'] = [50000, 60000, 70000]
print(df)
```

```python
# Group by 'Location' and calculate the average salary
grouped = df.groupby('Location')['Salary'].mean()
print(grouped)
```
